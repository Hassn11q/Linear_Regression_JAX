```python
import numpy as np
import matplotlib.pyplot as plt
#LinearREgression
# y = w*x+b
X = np.random.normal(size=(100,))
y = 3*X-1
plt.scatter(X,y)
```

```python
import jax.numpy as jnp
from jax import grad
def forward_pass(theta , x):
  w,b = theta
  return w*x+b
def loss_fun(theta , x , y):
  prediction = forward_pass(theta , x)
  return jnp.mean((prediction-y)**2)
```

```python
#init parameters
theta = jnp.array([0.,0.])
lr = 0.1

#gradient descent
for _ in range(1000):
  d_theta = grad(loss_fun)
  theta = theta-lr*d_theta(theta , X , y)
```

```python
plt.scatter(X,y)
pred = forward_pass(theta , X)
plt.plot(X,pred)
w,b = theta
print(f"W: {w:<.2f}, b: {b:<.2f}")
print(f"f(x)= {w:<.2f} * x {b:<.2f}")
```
